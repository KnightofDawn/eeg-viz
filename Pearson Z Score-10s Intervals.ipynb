```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
from scipy import stats
from scipy.io import loadmat
```

```python
m = loadmat('s5d2nap_justdata.mat')
matrix = np.array(m['s5d2nap'])
matrix
```

```python
channels_data = [[] for i in range(63)]
```

```python
# Z Scores 10 second intervals
start, end = 0, 10000
channels_data = [[] for i in range(63)]
while start <= len(matrix[0]):
    for channel in range(63):
        channels_data[channel].append(stats.zscore(matrix[channel][start:end]))
    start = end
    end += 10000
```

```python
pearson_5s = []
def pearson_(start):
    for channel in range(63):
#         print pearsonr(channels_data[start], channels_data[channel])
        if start < channel:
            for interval in range(len(channels_data[0])):
                pearson_5s.append(pearsonr(channels_data[start][interval], channels_data[channel][interval])[0])

for channel in range(63):
    print '%i ' % channel,
    pearson_(channel)
```

```python
# 0 line on graph
zero = []
for i in range(len(pearson_5s)):
    zero.append(0)

# plot data

# sorted
plt.plot(range(len(pearson_5s)), sorted(pearson_5s))
plt.plot(range(len(pearson_5s)), zero, 'k')


plt.title('Sorted Correlations between 2 Channels (5 second intervals) Pearson, Z')
plt.xlim(-1, len(pearson_5s) + 1)
plt.ylim(-1.1, 1.1)
plt.xlabel('Channel vs Channel')
plt.ylabel('Correlation value')
# plt.xticks(range(64), labels)
# plt.savefig('sorted_value_5second_64_z_score_pearson.svg')
plt.show()
```

```python
# abs value
# 0 line on graph
zero = []
for i in range(len(pearson_5s)):
    zero.append(0)

# plot data

# sorted
plt.plot(range(len(pearson_5s)), sorted(np.absolute(pearson_5s)))
plt.plot(range(len(pearson_5s)), zero, 'k')


plt.title('Sorted Absolute Correlations between 2 Channels (5 second intervals) Pearson, Z')
plt.xlim(-1, len(pearson_5s) + 1)
plt.ylim(0, 1.1)
plt.xlabel('Channel vs Channel')
plt.ylabel('Correlation value')
# plt.xticks(range(64), labels)
# plt.savefig('sorted_absolute_value_5second_64_z_score_pearson.svg')
plt.show()
```
