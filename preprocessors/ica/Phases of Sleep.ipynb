```python
def find_ts(timestep):
    if (timestep < 112):
        start = timestep * 30
        end = (timestep+1) * 30
        start_min_time = start/60
        start_sec_time = start%60
        end_min_time = end/60
        end_sec_time = end%60
        if start_sec_time <10:
            start_sec_time = "0" + str(start_sec_time)
        if end_sec_time < 10:
            end_sec_time = "0" + str(end_sec_time)

        start_time = str(start_min_time) + ":" + str(start_sec_time)
        end_time = str(end_min_time) + ":" + str(end_sec_time)
        print start_time + " - " + end_time
    elif (timestep == 112):
        print "56:00 - 56:24"
```

```python
for i in range(113):
    find_ts(i)
```
